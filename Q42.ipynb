**Questão 1.** No _dataset_ [enem2019.xlsx](https://github.com/gcpeixoto/ICD/blob/main/database/enem2019.xlsx), estão disponíveis as notas médias por estado obtidas nas provas do ENEM 2019. Supondo que _x_ é a diferença entre a amplitude da quantidade de inscritos na região Sudeste e a amplitude da quantidade de inscritos na região Norte, e que _y_ é o desvio médio para a série da quantidade total de inscritos de ensino médio público apenas para os estados do sul, assinale a alternativa que corretamente expressa os valores de _x_ e _y_,  nesta sequência.

**Obs.:** considere apenas a parte inteira do desvio médio.

***A. 149465 e 5690***

B. 169265 e 6593

C. 149465 e 0

D. 173921 e 2


<div class="alert alert-block alert-success">
    <b>Resposta:</b> A. 149465 e 5690
</div>

```python
# O primeiro dígito da coluna "COD UF" corresponde à região que aquele estado pertence, de tal maneira:
# 1 = Norte
# 2 = Noreste
# 3 = Sudeste
# 4 = Sul
# 5 = Centro-Oeste


import pandas as pd

df = pd.read_excel(
    "https://raw.githubusercontent.com/gcpeixoto/ICD/main/database/enem2019.xlsx"
)
df["COD UF"] = df["COD UF"].astype(str)


# Separando o dataframe original em dois:
df_norte = pd.DataFrame(columns=df.keys())
df_sudeste = pd.DataFrame(columns=df.keys())
df_sul = pd.DataFrame(columns=df.keys())
for i in range(len(df)):
    if df.at[i, "COD UF"].startswith("1"):
        df_norte = df_norte.append(df.iloc[i])
    elif df.at[i, "COD UF"].startswith("3"):
        df_sudeste = df_sudeste.append(df.iloc[i])
    elif df.at[i, "COD UF"].startswith("4"):
        df_sul = df_sul.append(df.iloc[i])
# Definindo as amplitudes
norte_amp   = df_norte['Quantidade de Inscritos'].max() - df_norte['Quantidade de Inscritos'].min()
sudeste_amp = df_sudeste['Quantidade de Inscritos'].max() - df_sudeste['Quantidade de Inscritos'].min()

# Garantindo que os elementos da série df_sul['Total de inscritos de Ensino Médio Público'] são números:
df_sul['Total de inscritos de Ensino Médio Público'] = df_sul['Total de inscritos de Ensino Médio Público'].astype(float)


# Respostas
x = sudeste_amp - norte_amp
y = df_sul['Total de inscritos de Ensino Médio Público'].mad()
```

```python
print(f"A diferença X entre a amplitude de inscritos nos estados do sudeste e a amplitude de inscritos nos estados do norte é de: {x}")
print(f"O desvio médio Y para a série da quantidade total de inscritos de ensino médio público apenas para os estados do sul é de: {y}")
```

**Questão 2.** Calcule o percentual _p_ de inscritos para o ENEM 2019 provenientes do ensino privado de todos os Estados em relação ao total de inscritos no exame, bem como o valor do quociente _v/V_, onde _v_ é a variância para a série do total de inscritos provenientes do ensino público e _V_ a variância para a série do total de inscritos provenientes do ensino privado. Assinale a alternativa correta para _p_ e _v/V_.

***A. 11.4% e 34.48***

B. 15% e 33.45

C. 12.5% e 36.78

D. 13.54% e 34.6

<div class="alert alert-block alert-success">
    <b>Resposta:</b> A. 11.4% e 34.48
</div>

```python
df = pd.read_excel(
    "https://raw.githubusercontent.com/gcpeixoto/ICD/main/database/enem2019.xlsx"
)

# Garantindo que são números e não objects
df['Total de inscritos de Ensino Médio Privado'] = df['Total de inscritos de Ensino Médio Privado'].astype(int)
df['Quantidade de Inscritos'] = df['Quantidade de Inscritos'].astype(int)

# o percentual p de inscritos para o ENEM 2019 provenientes do ensino
# privado de todos os Estados em relação ao total de inscritos no exame
p = df['Total de inscritos de Ensino Médio Privado'].sum()/df['Quantidade de Inscritos'].sum() * 100


# valor do quociente v/V, onde v é a variância para a série do total de inscritos provenientes
# do ensino público e V a variância para a série do total de inscritos provenientes do ensino privado
v = df['Total de inscritos de Ensino Médio Público'].var()
V = df['Total de inscritos de Ensino Médio Privado'].var()
```

```python
print(f"O valor para p é: {p}%")
print(f"A razão entre v e V é de: {v/V}")
```

**Questão 3.** Defina a nota média $N(x)$ de cada região brasileira $x$ como a média das notas $N_i$ de cada uma das $Q$ grandes áreas de conhecimento que constam da prova do ENEM 2019, isto é,

$$N(x) =  \frac{ \sum_{i=1}^Q N_i(x)}{Q},$$

e assinale a alternativa cujas regiões detém o primeiro e o segundo maiores valores de desvio padrão.

A. Nordeste e Sudeste

B. Sudeste e Nordeste

C. Norte e Sul

D. Sul e Centro-Oeste

```python
df = pd.read_excel(
    "https://raw.githubusercontent.com/gcpeixoto/ICD/main/database/enem2019.xlsx"
)
df["COD UF"] = df["COD UF"].astype(str)
# 1 = Norte
# 2 = Noreste
# 3 = Sudeste
# 4 = Sul
# 5 = Centro-Oeste
```

```python
# Defina a nota média  N(x)  de cada região brasileira  x  como a média das
# notas  Ni  de cada uma das  Q  grandes áreas de conhecimento
# que constam da prova do ENEM 2019, isto é,
```

```python
def definir_regiao(row):
    if row["COD UF"].startswith("1"):
        return "Norte"
    elif row["COD UF"].startswith("2"):
        return "Nordeste"
    elif row["COD UF"].startswith("3"):
        return "Sudeste"
    elif row["COD UF"].startswith("4"):
        return "Sul"
    else:
        return "Centro-Oeste"


# Criando a coluna médias
medias = []
for i in range(len(df)):
    soma = 0
    for key in df.keys():
        if "Média" in key:
            soma += df.at[i, key]
    medias.append(soma / 5)
df['Nota Média'] = medias
# Criando coluna Região
df["Região"] = df.apply(lambda row: definir_regiao(row), axis=1)
```
